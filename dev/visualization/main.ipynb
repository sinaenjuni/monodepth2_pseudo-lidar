```python
from pyntcloud import PyntCloud
import pandas as pd
import numpy as np


def paint_points(points, color=[192,0,0]):
    # color = [r, g, b]
    color = np.array([color])
    new_pts = np.zeros([points.shape[0],6])
    new_pts[:,:3] = points
    new_pts[:, 3:] = new_pts[:, 3:] + color
    return new_pts


# pd_lidar_points = pd.DataFrame(paint_points(lidar[:, :3], [192, 0, 0]), columns=['x','y','z','red','green','blue'])
# pd_lidar_points

# if __name__ == "__main__":

# file = np.fromfile("/Users/shinhyeonjun/Downloads/monodepth2-master/results/000000_pllidar.bin", dtype=np.float32)
# file = file.reshape((-1, 4))
# points = paint_points(file[:, :3], [255, 0, 0])
# pllidar_points = pd.DataFrame(points, columns=['x','y','z','red','green','blue'])

# file = np.fromfile("/Users/shinhyeonjun/Downloads/monodepth2-master/results/000000_pllidar_rectified.bin", dtype=np.float32)
file = np.fromfile("/Users/shinhyeonjun/Downloads/monodepth2-master/results/000000_pllidar_spartified.bin", dtype=np.float32)
file = file.reshape((-1, 4))
# file = spartify(file)
points = paint_points(file[:, :3], [0, 255, 0])
pllidar_spartified_points = pd.DataFrame(points, columns=['x','y','z','red','green','blue'])

# pllidar_spartified_points

file = np.fromfile("/Users/shinhyeonjun/dataset/data_object_velodyne/training/velodyne/000000.bin", dtype=np.float32)
file = file.reshape((-1, 4))
points = paint_points(file[:, :3], [0, 255, 255])
velo = pd.DataFrame(points, columns=['x','y','z','red','green','blue'])
# velo

pd_points = pd.concat((pllidar_spartified_points, velo))
pd_points
    
```
